```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams.update({
    "figure.dpi": 120,
    "figure.figsize": (10, 6)
})
```

```python
df1 = pd.read_csv('ghc_train.tsv', sep='\t')
df2 = pd.read_csv('ghc_test.tsv', sep='\t')

df = pd.concat([df1, df2], ignore_index=True)
df["class"] = np.where( (df["hd"] + df["cv"] + df["vo"]  > 0), "Hate", "Non-Hate") # If the row has at least one hate label, it is classified as Hate
len(df)
```

```python
# Cell 5: overall percentage of the two classes
class_percent = df["class"].value_counts()
colors = {"Hate":"red", "Non-Hate":"lightgreen"}

plt.figure()
class_percent.plot(kind="bar", color=[colors[c] for c in class_percent.index])
plt.title("Overall percentage per class")
plt.ylabel("Percentage (%)")
plt.xlabel("Class")
plt.xticks(rotation=0)
for i, v in enumerate(class_percent):
    plt.text(i, v + 0.7, v, ha="center")
plt.tight_layout()
plt.show()
```
