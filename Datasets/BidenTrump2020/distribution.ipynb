```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams.update({"figure.dpi": 120, "figure.figsize": (6, 4)})
```

```python
# Load train and test TSV files
train_df = pd.read_csv("train.tsv", sep="\t")
test_df = pd.read_csv("test.tsv", sep="\t")

# Combine
df = pd.concat([train_df, test_df], ignore_index=True)

print(f"Total rows {len(train_df)}+{len(test_df)}={len(df)}")
```

```python
# Count and normalize
class_percent = df["HOF"].value_counts(normalize=True) * 100
colors = {"Hateful": "red", "Non-Hateful": "lightgreen"}

plt.figure()
class_percent.plot(kind="bar", color=[colors.get(c, "gray") for c in class_percent.index])
plt.title("Percentage of classes (HOF column)")
plt.ylabel("Percentage (%)")
plt.xlabel("Class")
plt.xticks(rotation=0)

# Add percent labels
for i, val in enumerate(class_percent):
    plt.text(i, val + 0.5, f"{val:.1f}%", ha="center")

plt.tight_layout()
plt.show()
```
